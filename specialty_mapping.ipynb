```python
facilities_df = spark.table("databricks_virtue_foundation_dataset_dais_2026.virtue_foundation_dataset.facilities")

from pyspark.sql.functions import trim, lower, regexp_replace

cleaned_facebook_df = facilities_df.select(
    trim(regexp_replace("name", r"\s+", " ")).alias("hospital_name"),
    lower(trim(regexp_replace("facebookLink", r"\s+", ""))).alias("facebook_link")
).filter("facebook_link IS NOT NULL AND facebook_link != ''")
```
